```python
# CHAPTER 10 Data Aggregation and Group Operations
# GroupBy
# series.groupby([ndarray|series]) -> SeriesGroupBy
# df.groupby([column_name|ndarray|series]) -> DataFrameGroupBy

# GroupBy statistical method, 不统计缺失值
# GroupBy.mean()
# GroupBy.size()
# GroupBy.__iter__()

# split -> appply -> combine
# split:slice [:,:], logical and arithmetic 
# apply: pd.apply()
# combine: pd.concat()
import numpy as np
import pandas as pd
```

```python
# 10.1 GroupBy Mechanics

df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
                   'key2' : ['one', 'two', 'one', 'two', 'one'],
                   'data1' : np.random.randn(5),
                   'data2' : np.random.randn(5)})
df
```

```python
# series.groupby([ndarray|series]) -> SeriesGroupBy
grouped = df['data1'].groupby(df['key1'])
grouped
```

```python
# GroupBy.mean()
grouped.mean()
```

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means
```

```python
means.unstack()
```

```python
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
```

```python
df['data1'].groupby([states, years]).mean()
```

```python
# df.groupby([column_name|ndarray|series]) -> DataFrameGroupBy
df.groupby('key1')
```

```python
df.groupby('key1').mean()  # numerical data
```

```python
df.groupby(['key1', 'key2']).mean()
```

```python
# GroupBy.size()
df.groupby(['key1', 'key2']).size()  # exclude missing values in group
```

```python
# Iterating Over Groups
# GroupBy.__iter__()
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

```python
for name, group in df.groupby('key1').__iter__():
    print(name)
    print(group)
```

```python
for (k1, k2), group in df.groupby(['key1', 'key2']):
    print((k1, k2))
    print(group)
```

```python
# 10.3 Apply: General split-apply-combine
# split -> appply -> combine
# split:slice [:,:], logical and arithmetic 
# apply: pd.apply()
# combine: pd.concat()

def top(df, n=5, column='tip_pct'):
    return df.sort_values(by=column)[-n:]

top(tips, n=6)
```

```python
tips.groupby('smoker').apply(top)
```

```python
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
result.unstack('smoker')
```

```python
# Quantile and Bucket Analysis
frame = pd.DataFrame({'data1': np.random.randn(1000),
                     'data2': np.random.randn(1000)})
```

```python
# Example: Random Sampling and Permutation
# 模拟一副扑克牌52张及值
# Hearts, Spades, Clubs, Diamonds
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4 
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q'] 
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names) 
deck = pd.Series(card_val, index=cards)
```

```python
deck[:13]
```

```python
def draw(deck, n=5):
    return deck.sample(n)
draw(deck)
```

```python
get_suit = lambda card: card[-1] # last letter is suit
deck.groupby(get_suit).apply(draw, n=2)
```

```python
deck.groupby(get_suit, group_keys=False).apply(draw, n=2)
```

```python
rets = close_px.pct_change().dropna()
rets.head()
```

```python
spx_corr = lambda x: x.corrwith(x['SPX'])
get_year = lambda x: x.year
by_year = rets.groupby(get_year)
by_year.apply(spx_corr)
```

```python
# Example: Group-Wise Linear Regression
import statsmodels.api as sm 
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars] 
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit() 
    return result.params

by_year.apply(regress, 'AAPL', ['SPX'])
```
